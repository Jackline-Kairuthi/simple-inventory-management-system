#LIST ITEMS IN INVENTORY

```python
inventory = {'ginger biscuits': [22, 10.0, 15.82],
             'milk': [29, 33.0, 37.89],
             'sweets': [15, 2.09, 5.99],
             'sugar': [20, 15.78, 22.0],
             'bread': [15, 22.88, 30.0]}

#index 0 represents quantity
#index 1 represents base_price
#index 2 represents sale_price
```

#MINI PROGRAM TO ADJUST PRICE OR QUANTITY

```python
while True:
    print("\n--- MENU ---")
    print("1. Update quantity")
    print("2. Update price")
    print("3. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        product = input("Enter product name: ").lower()
        qty = int(input("Enter quantity to add: "))
        update_inventory(qty)
        print(inventory)

    elif choice == "2":
        product = input("Enter product name: ").lower()
        base = float(input("Enter amount to add to base price: "))
        sale = float(input("Enter amount to add to sale price: "))
        update_price(base, sale)
        print(inventory)

    elif choice == "3":
        print("Exiting program...")
        break

    else:
        print("Invalid choice. Try again.")
```

#UPDATE, QUANTITY AND PRICE OF ITEMS

```python
product = input("What is the name of the product: ")
def update_inventory(updated_quantity, new_base_price, new_sale_price):
  if product in inventory:
    inventory[product][0] += updated_quantity
    inventory[product][1] += new_base_price
    inventory[product][2] += new_sale_price
    return updated_quantity, new_base_price, new_sale_price
  else:
    print("product does not exist")

update_inventory(5,6,8)
print(inventory)
```

UPDATE ITEM QUANTITY

```python
product = input("What is the name of the product: ")
def update_inventory(updated_quantity):
  if product in inventory:
    inventory[product][0] += updated_quantity
    print("product already exists")
    print(f"updated quantity of product: {product}, is {inventory[product][0]}")

  else:
    print("product does not exist")

update_inventory(6)
print(inventory)
```

UPDATE ITEM PRICE

```python
product = input("What is the name of the product: ")
def update_price(updated_base_price, updated_sale_price):
  if product in inventory:
    inventory[product][1] += updated_base_price
    inventory[product][2] += updated_sale_price
    print("product already exists")
    print(f"updated prices of product: {product}, is {inventory[product][1]}, for base price and {inventory[product][2]} sale price")

  else:
    print("product does not exist")

update_price(2.5,6)
print(inventory)
```

#DISPLAY INVENTORY

# ADD ITEM

```python
product = input("What is the name of the product: ")

if product in inventory:
  print("product already exists")
  print(inventory[product])
else:
  quantity = int(input("What is the quantity of the product: "))
  base_price = float(input("What is the base price of the product: "))
  sale_price = float(input("What is the sale price of the product: "))
  inventory[product] = [quantity, base_price, sale_price]
  print(inventory)
```

#REMOVE PRODUCT

```python
product = input("What is the name of the product: ")

if product in inventory:
  del inventory[product]
  print("product deleted successfully")
else:
    print("product not found")

print(inventory)
```
